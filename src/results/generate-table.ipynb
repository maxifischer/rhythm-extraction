```python
import numpy as np
import pandas as pd

"""
x Alle Ergebnisse 
x beste features pro model, geordnet nach: bestes model ist oben
"""
```

```python
save_file_name = "merged.csv"
results = pd.read_csv(save_file_name)

results = results.dropna(subset=["cv_acc"])
results["param_linvar"].fillna(False, inplace=True)
results.columns
```

```python
results.round(decimals=3)

def print_float(x):
    if x==int(x):
        return str(int(x))+"."
    else:
        return "{0:.3f}".format(x)

formatters = {}
for col in ['cv_acc', 'test_acc', 'cv_f1', 'test_f1', "cv_acc_pc", "cv_acc_nc", "test_acc_pc", "test_acc_nc"]:
    formatters[col] = print_float

table = results[['data_name', 'prepr_name', 'model_name', 'cv_acc', 'test_acc', 'cv_f1', 'test_f1', "cv_acc_pc", "cv_acc_nc", "test_acc_pc", "test_acc_nc"]].groupby(
    ['data_name', 'model_name', 'prepr_name']).agg(max).groupby(level=[0,1], group_keys=False).apply(lambda x: x.sort_values('cv_acc', ascending=False))
table.groupby(level=[0, 1], group_keys=False).apply(lambda x: x.sort_values('cv_acc', ascending=False))
table.reset_index(level=0, inplace=True)
table
```

```python
model_pre = ["mv_nn", "mv_svm"]
a = results.groupby(["data_name", "prepr_name", "model_name"])["cv_acc"].idxmax()
print(a.values)

for row_id in results.groupby(["data_name", "prepr_name", "model_name"])["cv_acc"].idxmax().values:
    row = results.iloc[row_id]
    if row["model_name"] in ["mv_nn", "mv_svm"]:
        model_names.append(row["model_name"]+"--"+row["hyper_params"])
model_names
```

```python
gtzan = results.loc[results.data_name=="GTZAN"]
coltrain = results.loc[results.data_name!="GTZAN"]


g_grouped = gtzan.groupby(['model_name', 'prepr_name']).agg(
    {"cv_acc": 'max'}
).groupby(level=0, group_keys=False).apply(lambda x: x.sort_values('cv_acc', ascending=False))

c_grouped = coltrain.groupby(['model_name', 'prepr_name']).agg(
    {"cv_acc": 'max'}
).groupby(level=0, group_keys=False).apply(lambda x: x.sort_values('cv_acc', ascending=False))
c_grouped.rename(columns={"cv_acc": "cv_acc col."})
g_grouped.join(c_grouped, rsuffix=" col.")
```

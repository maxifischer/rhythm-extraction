```python
import numpy as np
import pandas as pd
```

```python
save_file_name = "merged.csv"
results = pd.read_csv(save_file_name)

results = results.dropna(subset=["cv_acc"])
results["param_linvar"].fillna(False, inplace=True)
```

```python
print(results.model_name.unique())
```

```python
res_grouped = results.groupby(['data_name', 'prepr_name', 'model_name'])

# print(res_grouped.idxmax())

best_ind_grouped = res_grouped['cv_acc'].idxmax()
results.iloc[best_ind_grouped].sort_values(['data_name', 'prepr_name', 'model_name'])
```
